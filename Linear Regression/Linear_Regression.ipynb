```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/placement.csv')

df.sample(6)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.title("CGPA vs Packages")
plt.xlabel(" CGPA ")
plt.ylabel("Packages (in LPA)")
plt.show()
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,1:2]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_yest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test.iloc[0].values.reshape(1,1)
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
y_yest
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_test['cgpa'],lr.predict(x_test),color='red')  ## best fit line
plt.title("CGPA vs Packages")
plt.xlabel(" CGPA ")
plt.ylabel("Packages (in LPA)")
plt.show()
```

```python
m=lr.coef_
b=lr.intercept_

##   ->   y = mx + b
print(f'm -> {m} , b -> {b}')
```
